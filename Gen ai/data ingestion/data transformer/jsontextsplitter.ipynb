```python
from sklearn.feature_extraction.text import CountVectorizer # Bag of Words (BoW)

documents = [
    "I love AI",
    "I love Machine Learning",
    "AI is amazing"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer # TF-IDF (Term Frequency-Inverse Document Frequency)

documents = [
    "I love AI",
    "I love Machine Learning",
    "AI is amazing"
]

tfidf = TfidfVectorizer()

X = tfidf.fit_transform(documents)

print(tfidf.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red"]
})

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df)

print(encoded)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data = ["Cat","Dog","Cat","Bird"]

result = encoder.fit_transform(data)

print(result)
```

```python
from sklearn.feature_extraction.text import CountVectorizer #N-Gram

documents = [
    "I love AI",
    "I love Python"
]

vectorizer = CountVectorizer(ngram_range=(2,2))

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
```
