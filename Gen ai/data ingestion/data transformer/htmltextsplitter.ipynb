```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string = """
<html>
<body>
<h1>Introduction</h1>
<p>This is the introduction section.</p>

<h2>Chapter 1</h2>
<p>Content of chapter 1.</p>

<h2>Chapter 2</h2>
<p>Content of chapter 2.</p>
</body>
</html>
"""
```

```python
headers_to_split_on = [("h1", "Header 1"), ("h2", "Header 2")]
```

```python
htmltext_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
```

```python
html_header_splits = htmltext_splitter.split_text(html_string)
```

```python
html_header_splits
```

```python
with open("sample.html", "r", encoding="utf-8") as file:
    html_string = file.read()
```
