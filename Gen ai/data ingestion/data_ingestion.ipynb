Chunking in LangChain

Chunking means breaking a large document into smaller pieces (chunks) so that LLMs can process them efficiently.

Why do we need chunking?

LLMs have token limits.
Large PDFs, books, or documents cannot be sent at once.
Smaller chunks improve retrieval accuracy in RAG systems.
