**Pinecone** is a fully managed **cloud-based vector database** designed specifically for Artificial Intelligence (AI) and Retrieval-Augmented Generation (RAG) applications.

It stores **vector embeddings** instead of traditional rows and columns and performs **semantic similarity search** to retrieve the most relevant information.

Unlike FAISS and ChromaDB, Pinecone stores your vectors in the **cloud**, making them accessible from anywhere.

# Similarity (Distance) Types in Vector Databases

When a user asks a question, the query is first converted into a vector (embedding).

The vector database then compares this query vector with all stored vectors and retrieves the most similar ones.

This comparison is done using a **Similarity (Distance) Metric**.

The most commonly used similarity metrics are:

1. Cosine Similarity
2. Euclidean Distance (L2 Distance)
3. Manhattan Distance (L1 Distance)
4. Dot Product (Inner Product)
5. Minkowski Distance
6. Hamming Distance
7. Jaccard Similarity
8. Pearson Correlation
9. Mahalanobis Distance
10. Chebyshev Distance

---

# 1. Cosine Similarity ⭐⭐⭐⭐⭐

## Definition

Cosine Similarity measures the **angle** between two vectors instead of their length.

Vectors pointing in the same direction are considered similar.

---

## Formula

\[
Cosine(A,B)=\frac{A\cdot B}{||A||||B||}
\]

---

## Range

```
-1 -------- 0 -------- 1
```

| Value | Meaning |
|--------|---------|
| 1 | Exactly Similar |
| 0 | No Similarity |
| -1 | Opposite |

---

## Example

```
Sentence 1

I love AI
```

```
Sentence 2

I enjoy Artificial Intelligence
```

Similarity

```
0.97
```

Very Similar

---

## Advantages

- Ignores vector magnitude
- Great for text embeddings
- Fast
- Most commonly used in RAG

---

## Used In

- Pinecone
- Chroma
- FAISS
- OpenAI Embeddings
- Hugging Face Embeddings

---

# 2. Euclidean Distance (L2 Distance)

## Definition

Measures the **straight-line distance** between two vectors.

---

## Formula

\[
d=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
\]

---

## Example

```
Point A

(2,3)

Point B

(5,7)
```

Distance

```
5
```

---

## Range

```
0 -------- ∞
```

Smaller distance

↓

More Similar

---

## Advantages

Easy to understand

Works well for numerical data

---

## Disadvantages

Affected by scale

---

## Used In

- KNN
- Clustering
- Image Processing

---

# 3. Manhattan Distance (L1 Distance)

## Definition

Measures the distance by moving only horizontally and vertically.

Also called

```
City Block Distance
```

---

Example

```
A

□────□────□
│
│
□
│
│
B
```

Cannot move diagonally.

---

Formula

\[
|x_1-x_2|+|y_1-y_2|
\]

---

Advantages

Works well for Grid Data

---

Used In

- GPS
- Robot Navigation
- Grid Search

---

# 4. Dot Product (Inner Product)

## Definition

Measures similarity by multiplying corresponding vector values.

---

Formula

\[
A\cdot B=\sum A_iB_i
\]

---

Example

```
A=[1,2]

B=[3,4]
```

Dot Product

```
1×3+2×4=11
```

---

Advantages

Very Fast

Preferred for normalized embeddings

---

Used In

- Pinecone
- FAISS
- Deep Learning

---

# 5. Minkowski Distance

Generalized distance metric.

Formula

\[
(\sum |x_i-y_i|^p)^{1/p}
\]

---

Special Cases

```
p=1

↓

Manhattan Distance
```

```
p=2

↓

Euclidean Distance
```

---

Used In

Machine Learning

KNN

---

# Interview Questions

### Q1. Which similarity metric is most commonly used in RAG?

**Answer:** Cosine Similarity because it measures semantic similarity based on vector direction and is ideal for text embeddings.

---

### Q2. What is the difference between Cosine Similarity and Euclidean Distance?

- **Cosine Similarity** compares the **direction (angle)** between vectors and ignores their magnitude.
- **Euclidean Distance** measures the **straight-line distance** between vectors and is affected by magnitude.

---

### Q3. Which metrics are supported by Pinecone?

- Cosine Similarity
- Dot Product (Inner Product)
- Euclidean Distance (L2)

# Summary

- **Cosine Similarity:** Best for text embeddings and RAG.
- **Euclidean Distance:** Best for numerical data.
- **Manhattan Distance:** Best for grid-based movement.
- **Dot Product:** Common in deep learning.
- **Minkowski Distance:** Generalization of L1 and L2.
- **Hamming Distance:** Binary/string comparison.
- **Jaccard Similarity:** Set comparison.
- **Pearson Correlation:** Statistical relationships.
- **Mahalanobis Distance:** Multivariate and outlier detection.
- **Chebyshev Distance:** Maximum coordinate difference.

````python
import os

from dotenv import load_dotenv

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

from pinecone import Pinecone, ServerlessSpec
from langchain_pinecone import PineconeVectorStore

from langchain_ollama import ChatOllama
````

````python
load_dotenv()

PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")
````

````python
pc = Pinecone(
    api_key=PINECONE_API_KEY
)
````

````python
index_name = "rishika-rag"
````

````python
if index_name not in pc.list_indexes().names():

    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",

        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )
````

````python
loader = TextLoader("hello.txt")

documents = loader.load()
````

````python
text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=30
)

documents = text_splitter.split_documents(documents)
````

````python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
````

````python
vectorstore = PineconeVectorStore.from_documents(
    documents=documents,
    embedding=embedding_model,
    index_name=index_name
)
````

````python
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k":3}
)
````

````python
llm = ChatOllama(
    model="llama3.2",
    temperature=0
)
````

````python
query = "What is my name?"

docs = retriever.invoke(query)
````

````python
context = "\n".join(
    [doc.page_content for doc in docs]
)
````

````python
prompt = f"""
You are a helpful AI assistant.

Answer ONLY using the context below.

Context:
{context}

Question:
{query}

Answer:
"""
````

````python
response = llm.invoke(prompt)

print(response.content)
````

````python
while True:

    query = input("Ask: ")

    if query.lower() == "exit":
        break

    docs = retriever.invoke(query)

    context = "\n".join(
        [doc.page_content for doc in docs]
    )

    prompt = f"""
You are a helpful AI assistant.

Answer ONLY using the context below.

Context:
{context}

Question:
{query}

Answer:
"""

    response = llm.invoke(prompt)

    print(response.content)
````
