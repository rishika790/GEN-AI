| Prompt Type                    | Purpose                                                     | Example                                                            |
| ------------------------------ | ----------------------------------------------------------- | ------------------------------------------------------------------ |
| **1. Zero-shot Prompt**        | Give only the task, no examples.                            | "Translate 'Hello' to French."                                     |
| **2. One-shot Prompt**         | Give one example before the task.                           | "Dog → Animal\nRose → Plant\nApple → ?"                            |
| **3. Few-shot Prompt**         | Give multiple examples to guide the AI.                     | Several Q&A examples, then ask a new question.                     |
| **4. Role Prompt**             | Assign a role to the AI.                                    | "Act as a senior Data Scientist."                                  |
| **5. Instruction Prompt**      | Give direct instructions.                                   | "Summarize this article in 100 words."                             |
| **6. Contextual Prompt**       | Provide background information.                             | "I'm a beginner in Python. Explain loops."                         |
| **7. Chain-of-Thought Prompt** | Ask the AI to reason through the problem before answering.  | "Solve this math problem step by step."                            |
| **8. Output Format Prompt**    | Specify the desired output format.                          | "Return the answer as a Markdown table."                           |
| **9. Persona Prompt**          | Ask the AI to respond in a particular personality or style. | "Explain AI like a friendly teacher."                              |
| **10. Multi-step Prompt**      | Break a complex task into sequential steps.                 | "First analyze the dataset, then build a model, then evaluate it." |


# Prompt Template (LangChain)

A **PromptTemplate** allows us to create reusable prompts with variables.

### Example

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["name"],
    template="Hello {name}, welcome to Upflairs!"
)

print(prompt.format(name="Rishika"))
```

Output

```text
Hello Rishika, welcome to Upflairs!
```

# Chat Prompt Template

A **ChatPromptTemplate** is used for chat-based Large Language Models (LLMs). It allows us to create prompts consisting of multiple messages such as **System**, **Human**, and **AI**.

### Example

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("human", "Explain Machine Learning in simple terms.")
])

messages = prompt.format_messages()

print(messages)
```

# Prompting in RAG (Retrieval-Augmented Generation)

## What is RAG Prompting?

In Retrieval-Augmented Generation (RAG), the prompt is created by combining:

- User Question
- Retrieved Documents
- Instructions
- Output Format

The LLM generates its answer only from the retrieved information instead of relying solely on its internal knowledge.

---
